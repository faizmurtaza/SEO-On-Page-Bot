```python

import time
import pandas as pd
import numpy as np
from datetime import datetime as dt,date,timedelta
from random import randint
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings('ignore')
from datetime import datetime
import requests
from datetime import datetime,date,timedelta

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.firefox.firefox_binary import FirefoxBinary
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
global driver
```

```python
%%time

d_11=date.today()
d_11=d_11.strftime("%Y-%m-%d")
print(d_11)

h_11 = datetime.today().time().strftime("%H:%M")
print(h_11)

keyword_list = ['Oneplus']
```

```python
%%time
driver = webdriver.Chrome(executable_path="chromedriver.exe")
driver.get('http://www.google.com/')
driver.maximize_window()
keyword2_=[]
urls = []
rank = []
landing_page = []

pages_to_scrape = 2


for key in keyword_list:
    driver.get('http://www.google.com/')
    try:
        driver.find_element_by_css_selector("body > div.L3eUgb > div.o3j99.ikrT4e.om7nvf > form > div:nth-child(1) > div.A8SBwf > div.RNNXgb > div > div.a4bIc > input").send_keys(key)
        time.sleep(randint(1,5))

        try:
            driver.find_element_by_css_selector('body > div.L3eUgb > div.o3j99.ikrT4e.om7nvf > form > div:nth-child(1) > div.A8SBwf > div.FPdoLc.lJ9FBc > center > input.gNO89b').click()
        except:
            driver.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[2]/div[2]/div[5]/center/input[1]').click()
        rank_ = 1 
        for i in range(pages_to_scrape):
            if i >= 1:
                driver.find_element_by_css_selector("#pnnext > span.SJajHc.NVbCr").click()
                time.sleep(randint(1,5))


            else:
                pass

            html=driver.page_source
            soup = BeautifulSoup(html, "html.parser")
            table = soup.findAll("div",{"class","TbwUpd NJjxre"})
            content = soup.findAll("div",{"class","yuRUbf"})
            time.sleep(randint(1,5))
        
            for idx, con in enumerate(table):
                landing_page.append(content[idx].find('a')['href'])
                rank.append(rank_)
                keyword2_.append(key)
                urls.append(con.text.split(" ")[0])
                print(content[idx].find('a')['href'],con.text.split(" ")[0])
                rank_ = rank_ + 1
                
    except:
        pass
    
data=pd.DataFrame(list(zip(keyword2_,urls,rank,landing_page)),columns=['Keyword','URL','Rank','Landing Page'])

data['Title'] = ''
data['Date'] = d_11
data['Hour'] = h_11

data
print("--------------")
print("Scrapping Done")
print("--------------")
print("--------------")
print("Now Scrapping Titles")
print("--------------")

for url in range(data.shape[0]):
    
    try:
        temp = data['Landing Page'].iloc[url]
        driver.get(temp)
        time.sleep(1)
        html=driver.page_source
        soup = BeautifulSoup(html, "html.parser")
        title = soup.find('title').text
        data['Title'].iloc[url] = title
        print(temp,title)
    except:
        pass

data.head()

```
